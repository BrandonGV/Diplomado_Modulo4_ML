```python
#Soluci√≥n
# -*- coding: utf-8 -*-
"""Taller_Regresi√≥n_Bicicletas_Completo.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1ABC123XYZ
"""

# ===================================================
# OBJETIVO: Aprender a aplicar regresi√≥n lineal para predecir el n√∫mero de bicicletas alquiladas
# ===================================================

# 1. INTRODUCCI√ìN
print("===================================================")
print("1. INTRODUCCI√ìN")
print("===================================================")

print("""
¬øQu√© tipo de problemas se pueden resolver con regresi√≥n lineal?

RESPUESTA: La regresi√≥n lineal se utiliza para problemas donde queremos predecir
un valor continuo (num√©rico) basado en una o m√°s variables independientes.

Ejemplos:
- Predecir precios de viviendas
- Estimar ventas de productos
- Pronosticar temperatura
- Predecir alquiler de bicicletas (como en este caso)
""")

# 2. CARGA DE DATASET
print("\n" + "="*50)
print("2. CARGA DE DATASET")
print("="*50)

# Importar librer√≠as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print("üì¶ Librer√≠as importadas correctamente")

# Descargar dataset
print("üì• Descargando dataset...")
!wget -q https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/daily-bike-share.csv

# Cargar datos
bike_data = pd.read_csv("daily-bike-share.csv")
print("‚úÖ Dataset cargado correctamente")

# Mostrar informaci√≥n del dataset
print(f"üìä Dataset shape: {bike_data.shape}")
print("\nüîç Primeras 5 filas:")
print(bike_data.head())

print("\nüìã Informaci√≥n del dataset:")
print(bike_data.info())

# Descripci√≥n de campos
print("\nüìñ DESCRIPCI√ìN DE CAMPOS:")
descripcion_campos = {
    'instant': 'ID del registro',
    'dteday': 'Fecha',
    'season': 'Temporada (1:primavera, 2:verano, 3:oto√±o, 4:invierno)',
    'yr': 'A√±o (0:2011, 1:2012)',
    'mnth': 'Mes (1 a 12)',
    'holiday': 'Si el d√≠a es festivo (1) o no (0)',
    'weekday': 'D√≠a de la semana (0 a 6)',
    'workingday': 'Si es d√≠a laboral (1) o no (0)',
    'weathersit': 'Situaci√≥n clim√°tica (1:bueno, 2:moderado, 3:malo)',
    'temp': 'Temperatura normalizada (0 a 1)',
    'atemp': 'Temperatura percibida normalizada (0 a 1)',
    'hum': 'Humedad normalizada (0 a 1)',
    'windspeed': 'Velocidad del viento normalizada (0 a 1)',
    'casual': 'N√∫mero de usuarios casuales',
    'registered': 'N√∫mero de usuarios registrados',
    'rentals': 'N√∫mero total de bicicletas alquiladas' # Corrected column name from 'cnt' to 'rentals'
}

for campo, descripcion in descripcion_campos.items():
    print(f"  {campo}: {descripcion}")

print(f"\nüéØ VARIABLE A PREDECIR: 'rentals' (N√∫mero total de bicicletas alquiladas)")

# Procesamiento de fecha
print("\nüìÖ Procesando fechas...")
bike_data['dteday'] = pd.to_datetime(bike_data['dteday'])
bike_data['day'] = bike_data['dteday'].dt.day

print("\nüìã Primeras 10 filas con d√≠a extra√≠do:")
print(bike_data[['dteday', 'day']].head(10))

# 3. PREPARACI√ìN DE DATOS
print("\n" + "="*50)
print("3. PREPARACI√ìN DE DATOS")
print("="*50)

# Definir caracter√≠sticas y variable objetivo
features = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday',
           'weathersit', 'temp', 'atemp', 'hum', 'windspeed', 'day']
target = 'rentals' # Corrected column name from 'cnt' to 'rentals'

print(f"üéØ Caracter√≠sticas: {features}")
print(f"üéØ Variable objetivo: {target}")

# Estad√≠sticas descriptivas
print("\nüìä ESTAD√çSTICAS DESCRIPTIVAS - CARACTER√çSTICAS:")
print(bike_data[features].describe())

print("\nüìä ESTAD√çSTICAS DESCRIPTIVAS - VARIABLE OBJETIVO:")
print(bike_data[target].describe())

# 4. GR√ÅFICAS DE DATOS
print("\n" + "="*50)
print("4. AN√ÅLISIS VISUAL DE DATOS")
print("="*50)

print("üìà Generando gr√°ficas...")

# Configurar estilo de gr√°ficas
plt.style.use('default')
fig = plt.figure(figsize=(20, 15))

# Gr√°fica 1: Temperatura vs Alquileres
plt.subplot(2, 3, 1)
plt.scatter(bike_data['temp'], bike_data['rentals'], alpha=0.6, color='blue') # Corrected column name from 'cnt' to 'rentals'
plt.xlabel('Temperatura Normalizada')
plt.ylabel('Bicicletas Alquiladas')
plt.title('Temperatura vs Alquileres')
plt.grid(True, alpha=0.3)

# Gr√°fica 2: Humedad vs Alquileres
plt.subplot(2, 3, 2)
plt.scatter(bike_data['hum'], bike_data['rentals'], alpha=0.6, color='green') # Corrected column name from 'cnt' to 'rentals'
plt.xlabel('Humedad Normalizada')
plt.ylabel('Bicicletas Alquiladas')
plt.title('Humedad vs Alquileres')
plt.grid(True, alpha=0.3)

# Gr√°fica 3: Velocidad del viento vs Alquileres
plt.subplot(2, 3, 3)
plt.scatter(bike_data['windspeed'], bike_data['rentals'], alpha=0.6, color='red') # Corrected column name from 'cnt' to 'rentals'
plt.xlabel('Velocidad del Viento Normalizada')
plt.ylabel('Bicicletas Alquiladas')
plt.title('Viento vs Alquileres')
plt.grid(True, alpha=0.3)

# Gr√°fica 4: Alquileres por estaci√≥n
plt.subplot(2, 3, 4)
season_means = bike_data.groupby('season')['rentals'].mean() # Corrected column name from 'cnt' to 'rentals'
season_names = ['Primavera', 'Verano', 'Oto√±o', 'Invierno']
plt.bar(season_names, season_means, color=['lightgreen', 'coral', 'gold', 'lightblue'])
plt.xlabel('Estaci√≥n')
plt.ylabel('Promedio de Alquileres')
plt.title('Alquileres Promedio por Estaci√≥n')
plt.grid(True, alpha=0.3)

# Gr√°fica 5: Alquileres por d√≠a de semana
plt.subplot(2, 3, 5)
weekday_means = bike_data.groupby('weekday')['rentals'].mean() # Corrected column name from 'cnt' to 'rentals'
weekday_names = ['Dom', 'Lun', 'Mar', 'Mi√©', 'Jue', 'Vie', 'S√°b']
plt.bar(weekday_names, weekday_means, color='purple', alpha=0.7)
plt.xlabel('D√≠a de la Semana')
plt.ylabel('Promedio de Alquileres')
plt.title('Alquileres por D√≠a de Semana')
plt.grid(True, alpha=0.3)

# Gr√°fica 6: Distribuci√≥n de alquileres
plt.subplot(2, 3, 6)
plt.hist(bike_data['rentals'], bins=30, color='orange', alpha=0.7, edgecolor='black') # Corrected column name from 'cnt' to 'rentals'
plt.xlabel('N√∫mero de Bicicletas Alquiladas')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n de Alquileres')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("üîç OBSERVACIONES:")
print("- Se observa relaci√≥n positiva entre temperatura y alquileres")
print("- La humedad y velocidad del viento tienen relaci√≥n negativa")
print("- Verano y oto√±o son las estaciones con m√°s alquileres")
print("- Los fines de semana tienen patrones diferentes de alquiler")

# 5. DIVISI√ìN EN CONJUNTOS DE ENTRENAMIENTO Y PRUEBA
print("\n" + "="*50)
print("5. DIVISI√ìN DE DATOS")
print("="*50)

X = bike_data[features]
y = bike_data[target]

# Dividir datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"üìä Tama√±o del conjunto de entrenamiento: {len(X_train)} filas ({len(X_train)/len(X)*100:.1f}%)")
print(f"üìä Tama√±o del conjunto de prueba: {len(X_test)} filas ({len(X_test)/len(X)*100:.1f}%)")

print("""
‚ùì ¬øQu√© pasa si usamos un test_size demasiado peque√±o?
RESPUESTA:
- La evaluaci√≥n del modelo puede no ser representativa
- Mayor riesgo de sobreajuste (overfitting)
- Menos confianza en la generalizaci√≥n del modelo
- No se captura adecuadamente la variabilidad de los datos
""")

# 6. ENTRENAMIENTO DEL MODELO
print("\n" + "="*50)
print("6. ENTRENAMIENTO DEL MODELO")
print("="*50)

print("ü§ñ Entrenando modelo de regresi√≥n lineal...")

# Crear y entrenar modelo
model = LinearRegression()
model.fit(X_train, y_train)

print("‚úÖ Modelo entrenado correctamente")

# Mostrar coeficientes
print("\nüìà COEFICIENTES DEL MODELO:")
for i, feature in enumerate(features):
    print(f"  {feature:12}: {model.coef_[i]:>8.2f}")

print(f"  {'Intercepto':12}: {model.intercept_:>8.2f}")

# Mostrar ecuaci√≥n
print("\nüßÆ ECUACI√ìN DEL MODELO:")
equation = f"alquileres = {model.intercept_:.2f}"
for i, coef in enumerate(model.coef_):
    if coef >= 0:
        equation += f" + {coef:.2f}√ó{features[i]}"
    else:
        equation += f" - {abs(coef):.2f}√ó{features[i]}"
print(equation)

print(f"\n‚ùì ¬øC√≥mo interpretas el coeficiente asociado a la temperatura?")
temp_coef_index = features.index('temp')
print(f"RESPUESTA: Por cada aumento de 1 unidad en temperatura normalizada,")
print(f"el n√∫mero de bicicletas alquiladas aumenta en {model.coef_[temp_coef_index]:.0f} unidades")

# 7. EVALUACI√ìN DEL MODELO
print("\n" + "="*50)
print("7. EVALUACI√ìN DEL MODELO")
print("="*50)

print("üìä Evaluando modelo...")

# Generar predicciones
y_pred = model.predict(X_test)

# Calcular m√©tricas
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("‚úÖ M√©tricas calculadas:")
print(f"  MSE  (Error Cuadr√°tico Medio): {mse:>10.2f}")
print(f"  RMSE (Ra√≠z del Error Cuadr√°tico Medio): {rmse:>10.2f}")
print(f"  MAE  (Error Absoluto Medio): {mae:>10.2f}")
print(f"  R¬≤   (Coeficiente de Determinaci√≥n): {r2:>10.4f}")

# Gr√°fica de predicciones vs reales
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.6, color='blue', label='Predicciones')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='L√≠nea perfecta')
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')
plt.title(f'Comparaci√≥n: Valores Reales vs Predicciones (R¬≤ = {r2:.3f})')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("""
üìã INTERPRETACI√ìN DE M√âTRICAS:

MSE: Mide el promedio de los errores al cuadrado. Valores m√°s bajos son mejores.
RMSE: Est√° en las mismas unidades que la variable objetivo.
      En promedio, nuestras predicciones se desv√≠an {:.0f} bicicletas del valor real.
R¬≤:  Indica qu√© porcentaje de la variabilidad en los datos es explicado por el modelo.
      Nuestro modelo explica el {:.1f}% de la variabilidad en los alquileres.
""".format(rmse, r2*100))

# 8. COMPARACI√ìN CON MODELO SIMPLE
print("\n" + "="*50)
print("8. COMPARACI√ìN DE MODELOS")
print("="*50)

print("üîç Entrenando modelo simple (solo con temperatura)...")

# Modelo solo con temperatura
X_temp = bike_data[['temp']]
y_temp = bike_data[target]

X_temp_train, X_temp_test, y_temp_train, y_temp_test = train_test_split(
    X_temp, y_temp, test_size=0.3, random_state=42
)

model_simple = LinearRegression()
model_simple.fit(X_temp_train, y_temp_train)
y_temp_pred = model_simple.predict(X_temp_test)

# M√©tricas del modelo simple
mse_simple = mean_squared_error(y_temp_test, y_temp_pred)
rmse_simple = np.sqrt(mse_simple)
r2_simple = r2_score(y_temp_test, y_temp_pred)

print("‚úÖ Modelo simple entrenado")

# Comparaci√≥n
print("\nüìä COMPARACI√ìN DE MODELOS:")
print("+" + "-"*50 + "+")
print("| {:25} | {:10} | {:10} |".format("M√âTRICA", "MULTIVARIABLE", "SIMPLE"))
print("+" + "-"*50 + "+")
print("| {:25} | {:10.2f} | {:10.2f} |".format("MSE", mse, mse_simple))
print("| {:25} | {:10.2f} | {:10.2f} |".format("RMSE", rmse, rmse_simple))
print("| {:25} | {:10.4f} | {:10.4f} |".format("R¬≤", r2, r2_simple))
print("+" + "-"*50 + "+")

# Gr√°fica comparativa
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Modelo multivariable
ax1.scatter(y_test, y_pred, alpha=0.6, color='blue')
ax1.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
ax1.set_xlabel('Valores Reales')
ax1.set_ylabel('Predicciones')
ax1.set_title(f'Modelo Multivariable\n(R¬≤ = {r2:.3f})')
ax1.grid(True, alpha=0.3)

# Modelo simple
ax2.scatter(y_temp_test, y_temp_pred, alpha=0.6, color='green')
ax2.plot([y_temp_test.min(), y_temp_test.max()], [y_temp_test.min(), y_temp_test.max()], 'r--', lw=2)
ax2.set_xlabel('Valores Reales')
ax2.set_ylabel('Predicciones')
ax2.set_title(f'Modelo Simple - Solo Temperatura\n(R¬≤ = {r2_simple:.3f})')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n‚ùì ¬øCu√°l modelo es m√°s preciso y por qu√©?")
if r2 > r2_simple:
    improvement = ((r2 - r2_simple) / r2_simple) * 100
    print(f"‚úÖ RESPUESTA: El modelo multivariable es m√°s preciso")
    print(f"   - Mejora la precisi√≥n en un {improvement:.1f}%")
    print(f"   - Considera m√∫ltiples factores que afectan los alquileres")
    print(f"   - Captura relaciones m√°s complejas entre variables")
else:
    print("‚ùå RESPUESTA: El modelo simple es m√°s preciso (resultado inesperado)")

# 9. CREACI√ìN DE DATASET SINT√âTICO
print("\n" + "="*50)
print("9. CREACI√ìN DE DATASET SINT√âTICO")
print("="*50)

print("üõ†Ô∏è Creando dataset sint√©tico para pr√°cticas futuras...")

np.random.seed(42)
n_samples = 1000

# Generar datos sint√©ticos
bike_dataset = pd.DataFrame({
    'season': np.random.choice([1, 2, 3, 4], n_samples, p=[0.2, 0.3, 0.3, 0.2]),
    'yr': np.random.choice([0, 1], n_samples, p=[0.5, 0.5]),
    'mnth': np.random.randint(1, 13, n_samples),
    'holiday': np.random.choice([0, 1], n_samples, p=[0.95, 0.05]),
    'weekday': np.random.randint(0, 7, n_samples),
    'workingday': np.random.choice([0, 1], n_samples, p=[0.3, 0.7]),
    'weathersit': np.random.choice([1, 2, 3], n_samples, p=[0.6, 0.3, 0.1]),
    'temp': np.random.normal(0.5, 0.2, n_samples).clip(0, 1),
    'atemp': np.random.normal(0.5, 0.2, n_samples).clip(0, 1),
    'hum': np.random.normal(0.6, 0.2, n_samples).clip(0, 1),
    'windspeed': np.random.normal(0.2, 0.1, n_samples).clip(0, 1),
    'day': np.random.randint(1, 32, n_samples)
})

# Calcular n√∫mero de bicicletas alquiladas basado en relaciones aprendidas
base_rentals = 2000
bike_dataset['rentals'] = ( # Corrected column name from 'cnt' to 'rentals'
    base_rentals +
    bike_dataset['season'] * 150 +
    bike_dataset['yr'] * 300 +
    bike_dataset['temp'] * 1000 +
    bike_dataset['atemp'] * 500 +
    -bike_dataset['hum'] * 300 +
    -bike_dataset['windspeed'] * 200 +
    -bike_dataset['weathersit'] * 100 +
    np.random.normal(0, 200, n_samples)
).clip(0, 8000)

# Mostrar estad√≠sticas del dataset creado
print("üìä DATASET SINT√âTICO CREADO - ESTAD√çSTICAS:")
print(bike_dataset.describe())

# Guardar dataset
bike_dataset.to_csv('bicicletas_dataset_sintetico.csv', index=False)
print("üíæ Dataset guardado como 'bicicletas_dataset_sintetico.csv'")

# Mostrar primeras filas
print("\nüîç Primeras 5 filas del dataset creado:")
print(bike_dataset.head())

# 10. PREDICCI√ìN CON DATOS NUEVOS
print("\n" + "="*50)
print("10. PREDICCI√ìN CON DATOS NUEVOS")
print("="*50)

print("üéØ Realizando predicciones con datos de ejemplo...")

# Datos de ejemplo para predicci√≥n
nuevos_datos = pd.DataFrame({
    'season': [2, 3, 1, 4],  # Verano, Oto√±o, Primavera, Invierno
    'yr': [1, 1, 0, 1],      # 2012, 2012, 2011, 2012
    'mnth': [7, 9, 4, 12],   # Julio, Septiembre, Abril, Diciembre
    'holiday': [0, 0, 1, 0], # No festivo, No festivo, Festivo, No festivo
    'weekday': [1, 5, 0, 3], # Lunes, Viernes, Domingo, Mi√©rcoles
    'workingday': [1, 1, 0, 1], # Laboral, Laboral, No laboral, Laboral
    'weathersit': [1, 2, 1, 3], # Bueno, Moderado, Bueno, Malo
    'temp': [0.8, 0.6, 0.5, 0.3], # Temperaturas
    'atemp': [0.75, 0.55, 0.45, 0.25], # Temperaturas percibidas
    'hum': [0.4, 0.6, 0.7, 0.8], # Humedad
    'windspeed': [0.1, 0.2, 0.3, 0.4], # Viento
    'day': [15, 20, 10, 25]  # D√≠as del mes
})

# Realizar predicciones
predicciones = model.predict(nuevos_datos)

print("üìã PREDICCIONES PARA DATOS NUEVOS:")
for i, pred in enumerate(predicciones):
    print(f"  Caso {i+1}: {pred:.0f} bicicletas alquiladas")

# Gr√°fica de predicciones
plt.figure(figsize=(10, 6))
casos = ['Verano\nLaboral', 'Oto√±o\nViernes', 'Primavera\nFestivo', 'Invierno\nLaboral']
plt.bar(casos, predicciones, color=['coral', 'gold', 'lightgreen', 'lightblue'])
plt.xlabel('Escenario')
plt.ylabel('Bicicletas Alquiladas Predichas')
plt.title('Predicciones de Alquiler para Diferentes Escenarios')
plt.grid(True, alpha=0.3)

# A√±adir valores en las barras
for i, v in enumerate(predicciones):
    plt.text(i, v + 50, f'{v:.0f}', ha='center', va='bottom')

plt.show()

print("\n" + "="*50)
print("üéâ TALLER COMPLETADO EXITOSAMENTE")
print("="*50)
print("""
‚úÖ Lo que hemos aprendido:
1. Carga y exploraci√≥n de datos
2. An√°lisis visual de relaciones
3. Preparaci√≥n de datos para ML
4. Entrenamiento de modelo de regresi√≥n lineal
5. Evaluaci√≥n con m√∫ltiples m√©tricas
6. Comparaci√≥n de modelos
7. Creaci√≥n de dataset sint√©tico
8. Predicciones con nuevos datos

üìÅ Archivos generados:
- bicicletas_dataset_sintetico.csv (dataset para pr√°cticas)
""")
```
